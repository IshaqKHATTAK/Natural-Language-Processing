```python
parameters = [".", "\n", ":", ",", "!", "?", " "]
```

```python
para = [".", "\n", ":", ",", "!", "?"]
```

```python
def Tokenize(sentences, i=0):
    test = []
    if i == len(para):
        test.append(sentences)
        return test
    try:
        flag = False
        if len(sentences.split(para[i])) > 1:
            flag = True
        for index, j in enumerate(sentences.split(para[i])):
            if flag == True:
                if(para[i] == " " or para[i] == " " or para[i] == "\n"):
                    pass
                elif(index%2 == 0):
                    j += para[i]
            test.extend(Tokenize(j, i+1))
    except:
        pass
    return test
```
