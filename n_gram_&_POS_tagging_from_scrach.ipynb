```python
import nltk
from collections import Counter
import matplotlib.pyplot as plt

# define the function for generating n-grams
def generate_ngrams(text, n):
    tokens = nltk.word_tokenize(text)
    ngrams = []
    for i in range(len(tokens)-(n-1)):
        ngram = tuple(tokens[i:i+n])
        ngrams.append(ngram)
    return ngrams

# define the text
text = "The quick brown fox jumps over the lazy dog."

# generate unigrams, bigrams, and trigrams
unigrams = generate_ngrams(text, 1)
bigrams = generate_ngrams(text, 2)
trigrams = generate_ngrams(text, 3)

# find the frequency of each n-gram
unigram_freq = Counter(unigrams)
bigram_freq = Counter(bigrams)
trigram_freq = Counter(trigrams)
```

# Pos tag

```python
# define the function for generating n-grams with POS tags
def generate_ngrams_with_pos(text, n):
    tokens = nltk.word_tokenize(text)
    tagged_tokens = nltk.pos_tag(tokens)
    ngrams = []
    for i in range(len(tagged_tokens)-(n-1)):
        ngram = tuple(tagged_tokens[i:i+n])
        ngrams.append(ngram)
    return ngrams

# generate bigrams with POS tags
bigrams_pos = generate_ngrams_with_pos(text, 2)

# find the frequency of each POS bigram
bigram_pos_freq = Counter(bigrams_pos)

# find the frequency of (NNP, VVP)
nnp_vvp_freq = 0
for bigram_pos in bigrams_pos:
    if bigram_pos[0][1] == 'NNP' and bigram_pos[1][1] == 'VVP':
        nnp_vvp_freq += 1

# plot the bar chart of POS tags
pos_tags = [tag for _, tag in nltk.pos_tag(nltk.word_tokenize(text))]
pos_tag_freq = Counter(pos_tags)
plt.bar(pos_tag_freq.keys(), pos_tag_freq.values())
plt.title("POS Tag Frequency")
plt.show()
```

```python
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import brown

# Load the shakespeare

text = nltk.corpus.gutenberg.raw("shakespeare-hamlet.txt")

# Generate N-grams, bi-grams and tri-grams using our code
tokens = word_tokenize(text)
n_grams = [tuple(tokens[i:i+n]) for n in range(1,4) for i in range(len(tokens)-n+1)]
n_gram_counts = Counter(n_grams)

# Generate N-grams using Sklearn CountVectorizer
vectorizer = CountVectorizer(ngram_range=(1,3))
X = vectorizer.fit_transform([text])
n_gram_counts_sk = Counter(dict(zip(vectorizer.get_feature_names_out(), X.toarray()[0])))

# Compare the results
print("Our code generated {} unique N-grams.".format(len(n_gram_counts)))
print("Sklearn CountVectorizer generated {} unique N-grams.".format(len(n_gram_counts_sk)))
print("The two N-gram counts match: {}".format(n_gram_counts == n_gram_counts_sk))
```
