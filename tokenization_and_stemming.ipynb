```python
# importing NLTK and different elements that we need in this file
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import gutenberg  # importing the corpus as data-set
```

```python
#required module
nltk.download('gutenberg')
nltk.download('punkt')
```

```python
sent = 'this is the > thiking asking ? for what i is khow nothing enable disable king'
```

```python
token = word_tokenize(sent)
print(token)
```

```python
stemm = PorterStemmer()

stm = [stemm.stem(word) for word in token]
print(stm)
```

###now try stemmer and tokenizer on nltk dataset

```python
hamlet = gutenberg.raw('shakespeare-hamlet.txt')  # converting it into a raw string
print(hamlet)
```

```python
tokens = word_tokenize(hamlet)  # performing tokenization
print(tokens)
```

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in tokens]  # performing lemmatization
print(stemmed_words)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = "Hello there, how are you doing today? something going too wrong The weather is great and python is awesome. The sky is pinkish-blue. You should not eat cardboard."
```

```python
from nltk.tokenize import sent_tokenize
sent_tok = sent_tokenize(text)
print(sent_tok)
```

```python
nltk.download('punkt')
```

```python
tok = word_tokenize(text)
print(tok)
```

```python
nltk.download('stopwords')
sw = stopwords.words('english')
filter = [word for word in tok if word not in sw]
print(filter)
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

sted = [ps.stem(word) for word in filter]
print(sted)
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')

wl = WordNetLemmatizer()
lmtize = [wl.lemmatize(word) for word in filter]
print(lmtize)
```

```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform([text])
print(bow.toarray())
```
