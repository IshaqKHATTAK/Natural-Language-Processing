```python
import nltk
from nltk.corpus import gutenberg
from nltk.tokenize import sent_tokenize
import re
```

```python
nltk.download('shakespeare')
nltk.download('punkt')
```

```python
st = 'this asking king getting let says. ? fro those of the kindforall disable enable U.S Army? for what i know him?'
```

```python
sent = sent_tokenize(st)
print(sent)
```

```python
token = [re.findall(r'\w+',word)  for word in sent]
print(token)
```

###now using the nltk dataset

```python
nltk.download('gutenberg')

shakespeare_text = gutenberg.raw('shakespeare-caesar.txt')

print(shakespeare_text)
```

```python
# Tokenizing the sentence using the nltk library

sentences = sent_tokenize(shakespeare_text)
print(sentences)
```

```python
# defining a regular expression to do a manual tokenization at words level for each sentence without using nltk.
token = [re.findall(r'\w+',sent) for sent in sentences]
print(token)
```
